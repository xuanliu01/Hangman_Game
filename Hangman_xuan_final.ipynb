```python
###################

import random

# Hangman dic
hangman={0:'''
        ____________
         |



         ''',
        1:'''
        ____________
         |
         O


         ''',
        2:'''
        ____________
         |
         O
        /

        ''',
        3:'''
        ____________
         |
         O
        / \\

        ''',
        4:'''
        ____________
         |
         O
        / \\
         |
         ''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}

word_dic = {"EASY": ["APPLE", "BREAD", "JUICE", "LEMON", "MANGO", "SAUCE"],
            "MEDIUM": ["RABBIT", "TURTLE", "CANDLE", "LAPTOP", "PUZZLE", "TICKET"],
            "HARD": ["MORNING", "KITCHEN", "PUDDING", "ORGANIC", "CARAMEL", "EVENING"]}

while True:
      difficulty = input("Choose the level of difficulty (easy, medium or hard): ").upper()
      if difficulty in word_dic:
        break
      else:
        print("Invalid input! Please enter easy, medium or hard.") #check invalid input

answer_word = random.choice(word_dic[difficulty])
hidden_list = ["_"] * len(answer_word)
hidden_word = " ".join(hidden_list)
print("=" * 60)  # Print a divider

remaining_mistakes = 6

while remaining_mistakes > 0 and "_" in hidden_word: # loop stop 1)loser 2)player wins

      print(f"Hidden word is:{hidden_word}")
      print("Pick a letter or guess a full word:")
      picked_letter = input().upper()

      ##check invalid input
      if not picked_letter.isalpha():
         print("Invalid input! Please enter letters only.")
         print("=" * 60)  # Print a divider
         continue

      if len(picked_letter) > 1:
          if picked_letter != answer_word:
             remaining_mistakes -= 1
             print(f"WRONG! Number of mistakes left: {remaining_mistakes}")
             print(hangman[6 - remaining_mistakes])  # Print the correct hangman stage
             print("=" * 60)  # Print a divider

          else:  # if picked_letter == answer_word
             hidden_word = " ".join(list(picked_letter))
             break

          continue

      if len(picked_letter) == 1:
         if picked_letter in answer_word:
            print(f"CORRECT! The word contains the letter {picked_letter}!")
            print("=" * 60)  # Print a divider
            #hidden_list = hidden_word.split() # ['_', '_', '_', '_', '_', '_']

            # replace the "_" with correct letters
            for index, letter in enumerate(answer_word):
                if letter == picked_letter:
                   hidden_list[index] = picked_letter # update correct position

                   # convert back to spaced out string
                   hidden_word = " ".join(hidden_list)

         else:
             remaining_mistakes -= 1
             print(f"WRONG! Number of mistakes left: {remaining_mistakes}")
             print(hangman[6 - remaining_mistakes])  # Print the correct hangman stage
             print("=" * 60)  # Print a divider

if remaining_mistakes == 0:
   print("HANGED!!")
else:
   print("CONGRATULATIONS!")

print(f"The correct word is {answer_word}!")
```
